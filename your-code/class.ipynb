```python
import requests
import pandas as pd

url = 'https://www.reuters.com/world/americas/more-officials-may-resign-compete-mexico-presidency-lopez-obrador-2023-06-07/'
html = requests.get(url).content
html[:600]
```

 we use BEAUTIFUL SOUP to clean the HTML we requested.

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(html, 'lxml')
soup
```

```python
#extracting specific tags

tags = ['h1', 'h2', 'h3', 'h4', 'h5', 'h6', 'h7', 'p']
text = [element.text for element in soup.find_all(tags)]
text
```

```python
#getting everything together

print('\n'.join(text))
```

## More complex Single-Page Scraping

```python
url = 'https://en.wikipedia.org/wiki/List_of_European_countries_by_life_expectancy'
html = requests.get(url).content
soup = BeautifulSoup(html)
soup
```

```python
table = soup.find_all('table', {'class': 'sortable wikitable'})[0]
table
```

```python
rows = table.find_all('tr')
rows = [rows.text.strip().split('\n') for rows in rows]
rows
```

```python
colnames = rows[0]
data = rows[1:]

df = pd.DataFrame(data, columns = colnames)
df
```
